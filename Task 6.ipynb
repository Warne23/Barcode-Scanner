```python
from pyzbar.pyzbar import decode
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def barcode_scanner(n):
    
    path = 'C:/Users/yugsi/Downloads/tasks/Task 6'
    
    image = cv2.imread(str(n)+'.jpg')

    original = image.copy()
    gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    blur=cv2.medianBlur(gray,5)

    thresh1 = cv2.Canny(blur,50,70)


    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
    dilate = cv2.dilate(thresh1, kernel, iterations=1)



    # Find contours, obtain bounding box coordinates, and extract ROI
    cnts,hierarchy = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
   
    

    image_number = 0


    large_contours = []

    for c in cnts:
        area_tresh = 10000
        area = cv2.contourArea(c)
        if area > area_tresh:
            cv2.drawContours(image, [c], contourIdx=-1, color=(0, 255, 255), thickness=1)
            large_contours.append(c)

   
    flag = 0

    for c in large_contours:
        
        x1,y1,w1,h1 = cv2.boundingRect(c)
        cv2.rectangle(image, (x1, y1), (x1 + w1, y1 + h1), (0,0,0), 10)
        ROI = original[y1:y1+h1, x1:x1+w1]

        detectedBarcodes = decode(ROI)

        if(len(detectedBarcodes)==0):
            continue
#             print('Not Detected')
#             cv2.rectangle(image, (x1, y1), (w1+x1, h1+y1), (0,0,255), 2)
#             print()

        else:
#             print('image_number = ',image_number+1)
#             print(detectedBarcodes)

            for barcode in detectedBarcodes:
                flag=1
                (x, y, w, h) = barcode.rect
#                 cv2.rectangle(image, (x+x1, y+y1), (x+x1 + w, y + h+y1), (255, 0, 0), 10)
#                 print(barcode.rect)
#                 print(barcode.data)
#                 print(barcode.type)
#                 print()
#         cv2.imwrite(os.path.join(path,("ROI_{}"+str(n)+'.png').format(image_number)), ROI)
            image_number += 1

    # cv2.imshow('image', image)

    
    
    cv2.imwrite(os.path.join(path , 'task_6_ '+str(n)+'.png'),image)
    cv2.waitKey()
```

```python
for i in range(4):
    barcode_scanner(i+1)
    barcode_scanner(str(i+1)+'missing')
    barcode_scanner(str(i+1)+'partial')
barcode_scanner('1miss_partial')
```
